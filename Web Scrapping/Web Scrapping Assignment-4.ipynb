# 1.List_of_most-viewed_YouTube_videos

```python
#webdriver address
driver=webdriver.Chrome(r"C:/Users/THIS PC/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
#Creating empty lists
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
#Extracting rank
try:
    rank=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in rank:
        Rank.append(i.text)
        
except NoSuchElementException:
    Rank.append("-")
    
#Extracting name
    
try:
    name=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]/a')
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
    
#Extracting artist
    
try:
    artist=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]/a')
    for i in artist:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
    
#Extracting views
    
try:
    view=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in view:
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")

#Extracting upload date    
    
try:
    upload=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in upload:
        Upload_date.append(i.text)

except NoSuchElementException:
    Upload_date.append("-")
    
    
    
    
    
```

```python
#Creating empty lists
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
#Extracting name of the repository
try:
    name=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a')
    for i in name:
        Dataset_name.append(i.text)
except NoSuchElementException:
    Dataset_name.append("-")
    
#Extracting data type    
try:
    types=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
    for i in types:
        Data_type.append(i.text)
except NoSuchElementException:
    Data_type.append("-")
    
#Extracting task    
try:
    task=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
    for i in name:
        Task.append(i.text)
except NoSuchElementException:
    Task.append("-")

#Extracting Attribute type    
try:
    attri=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
    for i in attri:
        Attribute_type.append(i.text)
except NoSuchElementException:
    Attribute_type.append("-")
    
#Extracting Number of instances    
try:
    insta=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')
    for i in insta:
        No_of_instances.append(i.text)
except NoSuchElementException:
    No_of_instances.append("-")
    
#Extracting number of attributes    
try:
    noa=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
    for i in noa:
        No_of_attribute.append(i.text)
except NoSuchElementException:
    No_of_attribute.append("-")
    
#Extracting year
try:
    year=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')
    for i in year:
        Year.append(i.text)
except NoSuchElementException:
    Year.append("-")    
```
