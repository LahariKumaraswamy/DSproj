```python
#Importing required libraies
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
import numpy as np
```

```python
#webdriver address
driver=webdriver.Chrome(r"C:/Users/THIS PC/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
#Getting amazon's url
driver.get('https://www.amazon.in')
```

```python
#Locating search bar
search_bar = driver.find_element_by_id("twotabsearchtextbox")    #Locating search bar throuh id
search_bar.clear()                                               #Clearing searhbox if there were any previous searches
```

```python
#Input
search_bar.send_keys(input("enter the produt"))
```

```python
#Locating search button
search_button = driver.find_element_by_xpath('//span[@id="nav-search-submit-text"]')#Locating searh button through xpath    
search_button.click()                                                               #Clicking searh button
```

```python
#Creating empty lists 
Brand_Name=[]
Name_of_the_product=[]
Rating=[]
Number_of_rating=[]
Price=[]
ReturnOrExchange=[]
Expected_delivery=[]
Availability=[]
Product_URL=[]
URL=[]
```

```python
#Gettinh all the urls
for i in range(1,3):
    urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")#collecting urls
    for i in urls:
        URL.append(i.get_attribute('href'))
        
        
    driver.find_element_by_xpath('//li[@class="a-last"]').click()  
    time.sleep(5)
```

```python
#Scraping each url indiviually
for url in URL:
    driver.get(url)
    time.sleep(10)
    
    #Extracting brand name
    try:
        brand=driver.find_element_by_xpath('//a[@id="bylineInfo"]')
        Brand_Name.append(brand.text)
    
    except NoSuchElementException:
           Brand_Name.append("-")
            
    #Extracting Name of the product
    try:
        nop=driver.find_element_by_xpath('//span[@id="productTitle"]')
        Name_of_the_product.append(nop.text)
    
    except NoSuchElementException:
           Name_of_the_product.append("-")
            
            
            
    #Extracting the price of the product
    try:
        price=driver.find_element_by_xpath('//span[@id="priceblock_ourprice"]')
        Price.append(price.text)

    except NoSuchElementException:
        Price.append("Not Available")
        
        
     #Extracting the product url

    try:
        produrl=driver.find_element_by_xpath('//a[@class="a-link-normal a-text-normal"]')
        Product_URL.append(produrl.get_attribute('href'))
    except NoSuchElementException:
        Product_URL.append('-')
        
     #Extracting rating and number of rating   
        
    try:                                                                        
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        nos=driver.find_element_by_xpath('//span[@class="a-size-base"]')
        Rating.append(rating.text)
        Number_of_rating.append(nos.text)
        
    except NoSuchElementException as e:
        Rating.append("-")
        Number_of_rating.append("-")
        
     #Extracting expected delivery   
        
    try:
        expd=driver.find_element_by_id("ddmDeliveryMessage")
        Expected_delivery.append(expd.text)
            
    except NoSuchElementException as e:
        Expected_delivery.append("-")
        
    #Extracting availability   
    try:
        stock=driver.find_element_by_xpath('//div[@id="availability"]/span')
        Availability.append(stock.text)
            
    except NoSuchElementException as e:
            Availability.append("-")
            
    #Extracting exchange        
            
    exchange=driver.find_elements_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')
    for i in exchange:
        if i.text=="30 Days Returns & Exchange":
            i.click()
            time.sleep(3)
            try:
                des=driver.find_element_by_xpath('//div[@class="a-section policy"]')
                ReturnOrExchange.append(des.text)
            
            except NoSuchElementException:
                     ReturnOrExchange.append("-")
```

```python
#Creating list of data
data=list(zip(Brand_Name,Name_of_the_product,Price,Product_URL,Rating,Number_of_rating,Expected_delivery,Availability,ReturnOrExchange))
```

```python
#Creating datafame
df=pd.DataFrame(data,columns=["Brand_Name","Name_of_the_product","Price","Product_URL","Rating","Number_of_rating","Expected_delivery","Availability","ReturnOrExchange"])
```

```python
#Setting index from 1
df.index = np.arange(1, len(df) + 1)
```

```python
driver.close()
```
